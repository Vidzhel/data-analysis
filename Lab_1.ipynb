```python
import csv

warehouse = [
  {'name': 'Apple', 'number': 200, "price": 1, "type": "fruits"},
  {'name': 'Banana', 'number': 150, "price": 2, "type": "fruits"},
  {'name': 'Avocado', 'number': 40, "price": 5, "type": "vegetables"},
  {'name': 'Potato', 'number': 530, "price": 2, "type": "vegetables"},
]

with open('warehouse.csv', 'wt', newline='') as frecord:
  crecord = csv.DictWriter(frecord, ['name', 'number', 'price', 'type'])
  crecord.writeheader()
  crecord.writerows(warehouse)
```
